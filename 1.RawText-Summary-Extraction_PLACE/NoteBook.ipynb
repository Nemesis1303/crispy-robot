```python
import pandas as pd
import os
from PIL import Image
from IPython.display import display
import glob
```

```python
filePath = '/export/data_ml4ds/thuban/repository/data/output/data.parquet'
```

```python
def openFile ( path ):
    try:
        data = pd.read_parquet(path)
        data.set_index('pdf_id',inplace=True)
        return data
        
    except Exception as E:
        print ('error abriendo el fichero %s: %s' % (path, E))
    
```

```python
data = openFile (filePath)
print ('se han cargado %s ficheros' % len(data.index))
```
