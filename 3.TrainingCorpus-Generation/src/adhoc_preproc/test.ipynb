```python
import pandas as pd
import pathlib
from collections import Counter
```

```python
path_data = pathlib.Path("/export/usuarios_ml4ds/lbartolome/thuban/crispy-robot/data/test.parquet")
```

```python
df = pd.read_parquet(path_data)
```

```python
# Define a function to filter NER tags
def filter_ners(ners, target_label):
    return [word for word, label in ners if label == target_label]

# Apply the function to the DataFrame
target_label = 'DRUG'  # Change this to the desired NER label
df[f'{target_label}_WORDS'] = df['raw_text_SPEC_NERS'].apply(lambda x: filter_ners(x, target_label))

# Display the DataFrame
df
```

```python
# Function to filter NER tags and keep words with their order and frequency
def filter_ners_with_order(row, target_label):
    words = row['raw_text_LEMMAS'].split()
    filtered_ners = [word for word, label in row['raw_text_SPEC_NERS'] if label == target_label]
    filtered_words = [word for word in words if word in filtered_ners]
    return ' '.join(filtered_words)

# Apply the function to the DataFrame
target_label = 'DRUG'  # Change this to the desired NER label
df[f'{target_label}_WORDS_ORDER'] = df.apply(lambda row: filter_ners_with_order(row, target_label), axis=1)
df
```

```python
def filter_ners_with_frequency(row, target_label):
    words = row['raw_text_LEMMAS'].split()
    word_freq = Counter(words)
    filtered_ners = [word for word, label in row['raw_text_SPEC_NERS'] if label == target_label]
    filtered_words = {word: word_freq[word] for word in filtered_ners if word in word_freq}
    return filtered_words

# Apply the function to the DataFrame
target_label = 'DRUG'  # Change this to the desired NER label
df[f'{target_label}_WORDS_ORDER'] = df.apply(lambda row: filter_ners_with_frequency(row, target_label), axis=1)
df
```

```python
# Function to filter rows based on the count of a specific NER label
def filter_rows_by_ner_count(df, target_label, min_count):
    def count_ners(ners, target_label):
        return sum(1 for _, label in ners if label == target_label)
    
    df['NER_COUNT'] = df['raw_text_SPEC_NERS'].apply(lambda x: count_ners(x, target_label))
    filtered_df = df[df['NER_COUNT'] >= min_count].copy()
    filtered_df.drop(columns=['NER_COUNT'], inplace=True)
    return filtered_df

# Specify the target NER label and minimum count
target_label = 'DRUG'  # Change this to the desired NER label
min_count = 2  # Minimum number of NER occurrences to keep the row

# Apply the function to filter the DataFrame
filtered_df = filter_rows_by_ner_count(df, target_label, min_count)
filtered_df
```
